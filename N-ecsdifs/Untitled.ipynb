```julia
using CSV
using Random
using DelimitedFiles
using DataFrames

x_value = rand(2)
total_1 = rand(2)
total_2 = rand(2)

# Write header to CSV file
CSV.write("data29.csv", DataFrame(Xs=[], Ys=[], Zs=[]), writeheader=true)

while true
    info = Dict(
        "Xs" => x_value,
        "Ys" => total_1,
        "Zs" => total_2
    )

    # Append data to CSV file
    CSV.write("data29.csv", DataFrame([info]), append=true, header=false)

    println("$x_value $total_1 $total_2")

    x_value += rand(2)
    total_1 += rand(2)
    total_2 += rand(2)

    sleep(1)
end
```
